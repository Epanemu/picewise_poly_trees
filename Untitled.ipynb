```python
n_features = 1
val_range = 100 # range of values of features [0, val_range]
n_data = 100    # samples used for the MIP solve 
n_sample = 1000 # sample for the ground truth visualization

def ground_truth(sample):
    x = sample[0]
    if x < 10:
        return x*2 + 2
    if x < 50:
        return x*4 - 100
    if x < 80:
        return x/2 + 20
    return -x*2 + 200

X = np.random.uniform(0, val_range, (n_data, n_features))
y = np.array([ground_truth(x) for x in X]) + np.random.normal(0, 10, (n_data,))
```

```python
dh = DataHandler(None, direct_data=(X, y, None, [f"x{i}" for i in range(n_features)], "Direct_data"))
X_train, y_train = dh.get_training_data(test_size=0)

# pwptree = PieceWisePolyTree_MIP(1, dh, poly_order=0)
# pwptree = PieceWisePolyTree_MIP(1, dh, poly_order=0, use_mse=True)
# pwptree = PieceWisePolyTree_MIP(1, dh, poly_order=1)
pwptree = PieceWisePolyTree_MIP(2, dh, poly_order=1)
# pwptree = PieceWisePolyTree_MIP(2, dh, poly_order=1, use_mse=True)
# pwptree = PieceWisePolyTree_MIP(3, dh, poly_order=1)

pwptree.make_model(X_train, y_train)

pwptree.optimize()
ctx = pwptree.get_base_context()
tree = Tree(ctx, dh)
```

```python
# X_space = (np.mgrid[0:val_range:val_range/n_sample, 0:val_range:val_range/n_sample]).T.reshape((n_sample**2, n_features))
X_space = np.linspace(0, val_range, n_sample).reshape((n_sample, n_features))

# print(X_space)
X_leaf = tree.get_assignment(X_space)

expo = pwptree.get_exponents(n_features)
leaf_values = dh.unnormalize_y(dh.normalize(X_space) @ expo @ ctx["poly_coeffs"] + ctx["intercepts"])
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(X_space, [ground_truth(x) for x in X_space], color="g")
ax.scatter(X[:, 0], y, color="r", marker="x")
for li in range(leaf_values.shape[1]):
    ax.plot(X_space[X_leaf==li, 0], leaf_values[X_leaf==li, li], color="b")
    ax.scatter(X_space[X_leaf==li, 0][0], leaf_values[X_leaf==li, li][0], color="b")
```

```python
my_pwlf = pwlf.PiecewiseLinFit(X.flatten(), y, degree=1)
res = my_pwlf.fit(7)
print(res)
pwlf_y = my_pwlf.predict(X_space.flatten())
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(X_space, [ground_truth(x) for x in X_space], color="g", label="Ground truth")
ax.scatter(X[:, 0], y, color="r", marker="x", label="data")
for li in range(leaf_values.shape[1]):
    ax.plot(X_space[X_leaf==li, 0], leaf_values[X_leaf==li, li], color="b", label="Our result" if li == 0 else None)
    ax.scatter(X_space[X_leaf==li, 0][0], leaf_values[X_leaf==li, li][0], color="b")
ax.plot(X_space, pwlf_y, "-.m", label="Pwlf result")
ax.legend()
plt.show()
```

# 3D variant

```python
n_features = 3
val_range = 100
n_data = 250

def ground_truth(sample):
    x, y, z = sample
    if x < 30:
        if y < 80:
            return x*4 - z
        else:
            if z > 40:
                return 2*y - 4*z
            return z
    if z < 50:
        return z/2 + x + y
    if y < 30:
        return 100
    return -x*2 + 200    

X = np.random.uniform(0, val_range, (n_data, n_features))
y = np.array([ground_truth(x) for x in X]) + np.random.normal(0, 10, (n_data,))
```

```python
n_sample = 20
xs, ys, zs = np.meshgrid(np.linspace(0, val_range, n_sample), np.linspace(0, val_range, n_sample), np.linspace(0, val_range, n_sample))
gt = np.zeros_like(xs)
for i in range(n_sample):
    for j in range(n_sample):
        for k in range(n_sample):
            gt[i,j,k] = ground_truth([xs[i,j,k], ys[i,j,k], zs[i,j,k]])

# fig = plt.figure()
# ax = fig.add_subplot(projection='3d')
# ax.scatter(xs, ys, zs, c=gt.flatten(), alpha=0.2)

# ax.set_xlabel('X')
# ax.set_ylabel('Y')
# ax.set_zlabel('Z')

# plt.show()
```

```python
dh = DataHandler(None, direct_data=(X, y, None, [f"x{i}" for i in range(n_features)], "Direct_data"))
X_train, y_train = dh.get_training_data(test_size=0)

# pwptree = PieceWisePolyTree_MIP(2, dh, poly_order=1)
pwptree = PieceWisePolyTree_MIP(3, dh, poly_order=1)

pwptree.make_model(X_train, y_train)

pwptree.optimize(time_limit=300) # 5 minutes max
ctx = pwptree.get_base_context()
tree = Tree(ctx, dh)
```

```python
X_space = np.concatenate([xs.reshape(-1,1), ys.reshape(-1,1), zs.reshape(-1, 1)], axis=1)

X_leaf = tree.get_assignment(X_space)

expo = pwptree.get_exponents(n_features)
leaf_values = dh.unnormalize_y(dh.normalize(X_space) @ expo @ ctx["poly_coeffs"] + ctx["intercepts"])

computed_vals = leaf_values[range(leaf_values.shape[0]), X_leaf].reshape((n_sample, n_sample, n_sample))

vmin = min(computed_vals.min(), gt.min())
vmax = max(computed_vals.max(), gt.max())

# fig, [ax, ax2] = plt.subplots(2, 1, figsize=(6,8), projection="3d")
fig = plt.figure(figsize=(6,8))
ax = fig.add_subplot(2, 1, 1, projection='3d')
ax.scatter(xs, ys, zs, c=gt.flatten(), alpha=0.2, edgecolors="none", vmin=vmin, vmax=vmax)
ax.set_title("Ground Truth")

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')


ax2 = fig.add_subplot(2, 1, 2, projection='3d')
cf2 = ax2.scatter(xs, ys, zs, c=computed_vals.flatten(), alpha=0.2, edgecolors="none", vmin=vmin, vmax=vmax)
# fig.colorbar(cf2, ax=[ax, ax2])
ax2.set_title("Computed")

ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Z')
plt.show()
```
